```python
import numpy as np
import matplotlib.pylab as plt
x = np.arange(-5.0, 5.0, 0.2)
global a,b,c
a = 2
b = 3
c = -3
#Defining the quadratic function
def f(x):
    return a*x**2+b*x+c

def fib(n):
    '''Return a Fibonacci series up to n'''
    a, b = 0, 1
    result=[0]
#Function adjusted to show n numbers
    while len(result)<n:
        result.append(b)
        a, b = b, a+b
    return result
```

```python
print(fib(20))
```
