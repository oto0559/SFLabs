## Exercise 1: Detemrining the ranges for the air resistance 

```python
import numpy as np
import matplotlib.pyplot as plt

B = 0.00016 #Constants
C = 0.25
D = 10**-3 #Diameter of the projectile in meters
b = B*D
c = C*D**2
g = 9.80665

def f(V):
    return b*V+c*V**2

vel = np.arange(0,0.0001,0.000001)
x = D*vel
y1 = b*vel
y2 = c*vel**2

plt.plot(x,y1, 'ro', label = 'bV')
plt.legend()
plt.show()
plt.plot(x,y2, 'bo', label = 'cV^2')
plt.legend()
plt.show()
```

Here it has been determined that for very low velocities and diameters (low D x V), the quadratic term becomes negligible
On the other hand, for big (D x V) the linear term is now negligible. In the middleground we need to consider both of these terms.

## Exercise 2: Vertical motion under air resistance

```python
rho = 2000 # Mass density
t = 0
dt = 0.01
Vy = 0
tmax = 50 # Maximum time of fall
m = rho * ((4*np.pi*(D/2)**3)/3)

Velocities = [0]
Times = [0]
Errors = [0]
t1 = np.arange(t,tmax+dt,dt)
def analytical(t1):
    return m*g/b*(np.exp(-b*t1/m)-1)

while t < tmax:
    t += dt
    Vy += -g*dt-(b/m)*Vy*dt
    Errors.append(analytical(t)-Vy)
    Times.append(t)
    Velocities.append(Vy)
    
plt.plot(Times,Velocities, 'go', ms = 3, label = 'Approximation')
plt.plot(t1,analytical(t1),'r--', ms = 2, label = 'Analytical solution')
plt.title('Free fall')
plt.xlabel('Time [s]')
plt.ylabel('Velocity [m/s]')
plt.legend()
plt.show()
plt.plot(Times,Errors, 'bo',ms=2)
plt.show()
```
