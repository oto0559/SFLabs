```python
import numpy as np
import matplotlib.pyplot as plt
global a,b,c
a = 2
b = 3
c = -3
def f(x):
    return a*x**2+b*x+c
def df(x):
    return 2*a*x+b
x = np.arange(-5.0, 5.0, 0.2)
plt.plot(x, f(x), label = f"${a}x^2+{b}x+{c}$")
plt.plot(x,df(x), label = f"${2*a}x+{b}$")
plt.plot(x, 0.0 * x, label = 'x=0')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
x1 = 1
x2 = -x1
x1 = x1 - f(x1)/df(x1)


tol = 0.0001
nsteps1 = 0
nsteps2 = 0
#Employing the Newton-Raphson method for the first root
while abs(f(x1)) > tol:
    nsteps1 += 1
    x1 = x1 - f(x1)/df(x1)
    plt.plot(x1,f(x1), 'ro')
#Printing first root
print('First root is:')    
print(f'x1 = {x1:.6f}')
print(f'The function value at that point is {f(x1):.10f}')
if f(x1) < tol:
    print(f'It took {nsteps1} steps to calculate the root within the given tolerance.')
#Newton-Raphson for second root

while abs(f(x2)) > tol:
    x2 = x2 - f(x2)/df(x2)
    plt.plot(x2,f(x2), 'yo')
    nsteps2 += 1
#Printing second root
print('Second root is:')    
print(f'x2 = {x2:.6f}')
print(f'The function value at that point is {f(x2):.10f}')
if f(x2) < tol:
    print(f'It took {nsteps2} steps to calculate the root within the given tolerance.')

plt.grid()
plt.xlim(-5,5)
plt.ylim(-5,7)
plt.show()
```

```python
#Loop for finding the root
steps = []
for i in range(-15, -1):
    x1 = 20
    x2 = -8
    tol = 10**(i)
    nsteps1 = 0
    nsteps2 = 0
    while abs(f(x1)) > tol:
        nsteps1 += 1
        x1 = x1 - f(x1)/df(x1)
    steps.append(nsteps1)
y = np.arange(-15,-1,1)
plt.plot(steps,y, 'bo')
plt.xlabel('Number of steps')
plt.ylabel('Log10 of tolerance')
plt.grid()
plt.show()
```
