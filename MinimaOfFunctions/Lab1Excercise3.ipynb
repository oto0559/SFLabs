```python
import numpy as np
import matplotlib.pyplot as plt
global const, A, p
const = 1.44
A = 1090
p = 0.033
x = np.arange(0.01,1,0.01)

def V(x):
    return A*np.exp(-x/p)-const/x
def dV(x):
    return (-A/p)*np.exp(-x/p)+const/x**2
def ddV(x):
    return (A/p**2)*np.exp(-x/p)-2*const/x**3

plt.plot(x,V(x), label = 'Potential energy [eV]')
plt.xlabel('Distance [nm]')
plt.ylabel('Interaction potential')

plt.plot(x,-dV(x), label='Force [eV/nm]')

plt.plot(x,0*x, 'black', linewidth=1)
plt.ylim(-20,100)

plt.legend()
plt.grid()
plt.show()
```

```python
def ddV(x):
    return (A/p**2)*np.exp(-x/p)-2*const/x**3
x1 = 0.2
tol = 0.00000001
nsteps=0
while abs(dV(x1)) > tol:
    nsteps+=1
    x1 = x1 - dV(x1)/ddV(x1)
print(x1)
print(dV(x1))
print(f'The minimum of V(x) is at x = {x1:.6f} nm and the number of steps is: {nsteps}')
```
